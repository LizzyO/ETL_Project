```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/IMDb_ratings.csv"
rating_data_df = pd.read_csv(csv_file)

movie_csv_file = "Resources/IMDb_movies.csv"
imdb_movies_df = pd.read_csv(movie_csv_file)

disney_csv_file = "Resources/disney_plus_shows.csv"
disney_data_df = pd.read_csv(disney_csv_file)

netflix_csv_file = "Resources/netflix_titles.csv"
netflix_data_df = pd.read_csv(netflix_csv_file)
```

# filter columns

```python
new_rating_data_df = rating_data_df[['imdb_title_id','weighted_average_vote']].dropna()
```

```python
imdb_movies_drop_df=imdb_movies_df[['imdb_title_id','title']].dropna()
new_imdb_movies_df = imdb_movies_drop_df.drop_duplicates(['title'])
```

```python
new_disney_data_df = disney_data_df[['imdb_id','title','imdb_rating','plot','type','year','added_at']].dropna()
disney_filter = new_disney_data_df.loc[new_disney_data_df['type']=='movie'].rename(columns={'imdb_id':'imdb_title_id'})
disney_filter
```

```python
new_netflix_data_df = netflix_data_df[['show_id','title','description','type','release_year','date_added']].rename(columns={'release_year':'year'})
netflix_filter = new_netflix_data_df.loc[netflix_data_df['type']=='Movie'].drop_duplicates(['title'])
netflix_filter
```

```python
conn = "postgres:postgres@localhost:5432/ETL_Project"
engine = create_engine(f'postgresql://{conn}')
```

```python
engine.table_names()
#new_rating_data_df
#new_imdb_movies_df
#disney_filter
#netflix_filter
```

```python
new_rating_data_df.to_sql(name='imdb_ratings',con=engine,if_exists='append',index=False)
new_imdb_movies_df.to_sql(name='imdb_title',con=engine,if_exists='append',index=False)
disney_filter.to_sql(name='disney',con=engine,if_exists='append',index=False)
netflix_filter.to_sql(name='netflix',con=engine,if_exists='append',index=False)
```
